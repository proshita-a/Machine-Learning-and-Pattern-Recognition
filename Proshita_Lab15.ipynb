```python
# step 1

from numpy import mean
from numpy import std
from numpy import dstack

from pandas import read_csv

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import Dropout
from keras.layers import LSTM

from keras.utils import to_categorical

from matplotlib import pyplot

import numpy as np
```

```python
# Load a single file as a numpy array

def load_file(filepath):

    dataframe = read_csv(
        filepath,
        header=None,
        delim_whitespace=True
    )

    return dataframe.values
```

```python
# Load all signal files

def load_group(filenames, prefix=''):

    loaded = list()

    for name in filenames:

        data = load_file(prefix + name)

        loaded.append(data)

    loaded = dstack(loaded)

    return loaded
```

```python
# Load dataset group (train or test)

def load_dataset_group(group, prefix=''):

    filepath = prefix + group + '/Inertial Signals/'

    filenames = list()

    filenames += ['total_acc_x_' + group + '.txt']
    filenames += ['total_acc_y_' + group + '.txt']
    filenames += ['total_acc_z_' + group + '.txt']

    filenames += ['body_acc_x_' + group + '.txt']
    filenames += ['body_acc_y_' + group + '.txt']
    filenames += ['body_acc_z_' + group + '.txt']

    filenames += ['body_gyro_x_' + group + '.txt']
    filenames += ['body_gyro_y_' + group + '.txt']
    filenames += ['body_gyro_z_' + group + '.txt']

    X = load_group(filenames, filepath)

    y = load_file(prefix + group + '/y_' + group + '.txt')

    return X, y
```

```python
# Load train and test dataset

def load_dataset(prefix='UCI HAR Dataset1/'):

    trainX, trainy = load_dataset_group(
        'train',
        prefix
    )

    testX, testy = load_dataset_group(
        'test',
        prefix
    )

    trainy = trainy - 1
    testy = testy - 1

    trainy = to_categorical(trainy)
    testy = to_categorical(testy)

    print(trainX.shape, trainy.shape)

    print(testX.shape, testy.shape)

    return trainX, trainy, testX, testy
```

```python
trainX, trainy, testX, testy = load_dataset()
```

```python
def evaluate_model(trainX, trainy, testX, testy):

    verbose = 1
    epochs = 15
    batch_size = 64

    n_timesteps = trainX.shape[1]
    n_features = trainX.shape[2]
    n_outputs = trainy.shape[1]

    model = Sequential()

    model.add(
        LSTM(
            100,
            input_shape=(n_timesteps, n_features)
        )
    )

    model.add(Dropout(0.5))

    model.add(Dense(100, activation='relu'))

    model.add(Dense(n_outputs, activation='softmax'))

    model.compile(
        loss='categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )

    history = model.fit(
        trainX,
        trainy,
        epochs=epochs,
        batch_size=batch_size,
        verbose=verbose
    )

    loss, accuracy = model.evaluate(
        testX,
        testy,
        batch_size=batch_size,
        verbose=0
    )

    return accuracy
```

```python
scores = list()

for r in range(10):

    score = evaluate_model(
        trainX,
        trainy,
        testX,
        testy
    )

    score = score * 100.0

    print('>#%d: %.3f' % (r + 1, score))

    scores.append(score)
```

```python
m = mean(scores)
s = std(scores)

print('Mean Accuracy: %.3f%% (+/- %.3f)' % (m, s))
```

### 1. Why was this problem suitable for LSTM and not just any other machine learning algorithm?

This problem was suitable for LSTM because the smartphone sensor data is sequential in nature. The activities are recorded over time, so the order of the data matters. LSTMs are specially designed to learn patterns from time-series and sequential data, unlike normal machine learning algorithms which usually treat all inputs independently.



### 2. Is a many-to-many model a more real-world scenario for this problem or a many-to-one model and why?

A many-to-many model would be a more real-world scenario because human activities continuously change over time. In real life, a person may go from walking to sitting or standing, so predicting activities at multiple time steps is more practical. However, in this assignment, a many-to-one model is used because one activity label is predicted for an entire sequence window.



### 3. Explain the meaning of вЂњShort MemoryвЂќ and вЂњLong MemoryвЂќ in context of LSTMs.

Short memory refers to information that is useful only for a short duration or a few time steps. Long memory refers to information that needs to be remembered for a longer sequence of time steps. LSTMs are designed to handle both types of memory using gates that decide what information to keep or forget.



### 4. What are the three parts of an LSTM node and their functions?

1. Forget Gate вЂ“ Decides which old information should be removed from memory.

2. Input Gate вЂ“ Decides which new information should be added to the memory.

3. Output Gate вЂ“ Decides what information should be passed to the next hidden state or output.



### 5. What problems do RNNs face that LSTMs are able to adapt to fix?

Traditional RNNs face the vanishing gradient problem, where important information from earlier time steps gets lost during training. Because of this, RNNs struggle to learn long-term dependencies. LSTMs solve this problem using memory cells and gating mechanisms that help preserve important information for longer durations.
